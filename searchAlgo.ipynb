```python

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
embedder = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
```

```python
data['embeddings'] = data['Unnamed: 1'].apply(lambda x: embedder.encode(x))
```

```python
import heapq
```

```python
input = "Humboldt Bay"
```

```python
heap = []
```

```python
embeddings = data['embeddings'].values
```

```python
names = data['Unnamed: 1'].values
```

```python
names
```

```python
testEmb = embedder.encode(input)
```

```python
def find_similar():
  for i in range(len(embeddings)):
    sim = cosine_similarity([testEmb], [embeddings[i]])
    heapq.heappush(heap, (sim, names[i]))

  return heapq.nlargest(20, heap)
```

```python
test = find_similar()
```

```python
import math
def time_optimal_find_similar(input):
    heap = []
    testToken = v.transform([input]).toarray()[0].reshape(1, -1)
    testEmb = embedder.encode(input)
    words = input.split()
    for i in range(len(tokens)):
        sim1 = cosine_similarity([testEmb], [embeddings[i]])
        add = 0
        mult = 1
        for word in words:
            if word in names[i]:
                add += 1 * mult
            mult = mult / 4
        sim2 = cosine_similarity(testToken, tokens[i].reshape(1, -1))
        heapq.heappush(heap, ((sim1 * 2.5) + add + (sim2 * 1.5), names[i]))

    return heapq.nlargest(50, heap)
```
